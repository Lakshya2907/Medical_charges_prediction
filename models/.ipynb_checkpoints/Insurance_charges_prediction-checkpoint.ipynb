```python
def rmse(targets,predictions):
    return np.sqrt(np.mean(np.square(predictions - targets)))
```

```python
from sklearn.metrics import r2_score
```

```python
non_smoker_df = df[df.smoker == 'no']
```

```python
ages = non_smoker_df.age
target = non_smoker_df.charges
```

### Function for best fit line rmse score and r2 score

```python
def estimate_charges(age, w, b):
    return w * age + b
```

```python
def try_parameters(w, b):
    predictions = estimate_charges(ages, w, b)
    
    plt.plot(ages, predictions, 'r', alpha=0.9);
    plt.scatter(ages, target, s=8,alpha=0.8);
    plt.xlabel('Age');
    plt.ylabel('Charges')
    plt.legend(['Prediction', 'Actual']);
    
    rmse_loss = rmse(target, predictions)
    r2 = r2_score(target, predictions)
    print("RMSE Loss: ", rmse_loss)
    print("R^2 score:", r2)
```

## Performing one hot encoding for region 

```python
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()
enc.fit(df[['region']])
enc.categories_
```

```python
one_hot = enc.transform(df[['region']]).toarray()
one_hot
```

```python
df[['northeast', 'northwest', 'southeast', 'southwest']] = one_hot
```

```python
df
```

```python
input_cols = ['age', 'bmi', 'children', 'smoker_codes', 'sex_codes', 'northeast', 'northwest', 'southeast', 'southwest']
inputs = df[input_cols]
target = df['charges']
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
numeric_col = ['age','bmi','children']
scaler.fit(df[numeric_col])
```

```python
print(scaler.mean_)
print(scaler.var_)
```

```python
scaled_inputs = inputs.copy()
scaled_inputs = scaler.transform(df[numeric_col])
scaled_inputs
```

```python
categorical_cols = ['sex_codes' , 'smoker_codes' , 'northeast', 'northwest', 'southeast', 'southwest']
categorical_data = df[categorical_cols]
categorical_data
```

## Combining categorical and numerical data

```python
inputs = np.concatenate((scaled_inputs, categorical_data), axis=1)
targets = df['charges']
inputs = pd.DataFrame(inputs)
inputs.rename(columns={0: 'age', 1: 'bmi',2: 'children',3: 'smoker_code',4: 'sex_code', 5: 'northeast',6: 'northwest',7: 'southeast',8: 'southwest'}, inplace=True)
inputs
```

## Linear Regression on whole data

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_test, targets_train, targets_test = train_test_split(inputs, targets, test_size=0.2,random_state = 0)
```

```python
inputs_train
```

```python
inputs_test
```

```python
model = LinearRegression()
model.fit(inputs_train, targets_train)
predictions_test = model.predict(inputs_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
rmse_loss = rmse(targets_test, predictions_test)
r2 = r2_score(targets_test, predictions_test)
print('Rmse Loss:', rmse_loss)
print('r2 score:', r2)
```
